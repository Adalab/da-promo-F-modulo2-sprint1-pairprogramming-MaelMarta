```python
diccionario_renovables_no_renovables = {"tipo_energia": [],
                             "tipo_generacion":[],
                             "fecha": [],
                             "valor": []}
meses = 0

for energias in json_generation["included"]: #Iteramos por cada energía
    
    for meses in range(10): #12 meses --> Me da error fuera de rango
        tipo_energia = energias["type"]
        tipo_generacion = energias["attributes"]["type"]
        
        #Hemos hecho una lista dentro de cada value
        diccionario_renovables_no_renovables["tipo_energia"].append(tipo_energia)
        
        diccionario_renovables_no_renovables["tipo_generacion"].append(tipo_generacion)
    
        print(energias["attributes"]["values"])
        fecha = energias["attributes"]["values"][meses]["datetime"]
        
        diccionario_renovables_no_renovables["fecha"].append(fecha)
        
        valor = energias["attributes"]["values"][meses]["value"] 
        diccionario_renovables_no_renovables["valor"].append(valor)


```
